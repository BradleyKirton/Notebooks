```python
# This module contains functions which we will use to pull data from google finance
import pandas.io.data as web
```

```python
# I found the constituents of the Satrix 40 exchange traded fund online. 
# We will use this reference data to query the google finance api with pandas
constituents = pd.read_csv('Satrix 40 ETF Basket Constituents.csv')

# Here we are just doing some data cleaning
constituents.Weight.replace('[^.\d]', '', regex=True, inplace=True)
constituents['Market cap'].replace('[^.\d]', '', regex=True, inplace=True)

constituents.Weight = constituents.Weight.astype(np.float)
constituents['Market cap'] = constituents['Market cap'].astype(np.float)

# The info method displays information about our data like data types, nulls and counts as well as an approximation of 
# the memory footprint.
constituents.info()
```

<p>So we have a number of fields in this file but what do they all mean?</p>

<ol>
    <li>The share code refers to a three (sometimes four) letter code the JSE asigns to each instrument that trades on its markets</li>

<li>The share name refers to the name of the listed product. In the case of our data above, all these products represent ordinary shares.</li>

<li>The weight field refers to the weight that the corresponding stock contributes to the portfolio. In our case this portfolio is designed to track the top 40 stocks on the JSE. We call them passive tracker funds.</li>

<li>Market capitalisation is a measure of how large a company is. It is calculated as the number of shares in issue multiplied by the shares price. The top 40 index happens to be market cap weighted, so the higher the market cap the larger the weight a stock would have in the index.</li>

<li>This closing price is the price of the stock on the day the constituents file was downloaded.</li>

<li>The industry refers to the industry of the issuing company.</li>

</ol>

```python
# As shown above we can have a quick look at our data by calling head
constituents.head()
```

```python
# Panads has a nice interface for plotting your data. Below we plot the weights per constituent
constituents.plot.bar(
    x='Share code',
    y='Weight',
    figsize=(12, 7)
)
```

```python
# One of the most powerful features of pandas is it's aggregation framework. 
# http://pandas.pydata.org/pandas-docs/stable/groupby.html

# Here we sum the weights per industry
constituents.groupby('Industry').Weight.sum()
```

```python
# So let's grab the data for the constituents for 2015 and calculate the return on a portfolio
constituent_list = ('JSE:' + constituents['Share code']).tolist()

panel = web.DataReader(constituent_list, 'google', '2016-02-25', '2016-02-25')

close = panel.Close
close.dropna(axis=1, inplace=True)
```

```python
# So here we can see a nice tabular set of closing prices per stock
close.head()
```

```python
# Here we calculate the cumulative return per stock for the period
rets = (close.pct_change() + 1).cumprod()[-1:].T - 1

rets.head()
```

```python
# Here we use the returns we calculated to find the worst 5 performers for 2015 and plot them
worst_performers = rets.sort_values(by='2015-12-31 00:00:00').head().index

worst = close[worst_performers]
worst_rebased = DataFrame(worst.values / worst.values[:1], index=worst.index, columns=worst.columns) * 100

worst_rebased.plot(
    figsize=(12, 7)
)
```

```python
# Here we use the returns we calculated to find the best 5 performers for 2015
best_performers = rets.sort_values(by='2015-12-31 00:00:00', ascending=False).head().index

best = close[best_performers]
best_rebased = DataFrame(best.values / best.values[:1], index=best.index, columns=best.columns) * 100

best_rebased.plot(
    figsize=(12, 7)
)
```

<h2>Industry Neutral Portfolio</h2>

```python
industry_count = constituents.Industry.value_counts()
industry_count = industry_count.to_frame('Weight')
```

```python
# Below the weights per industry are calculated. 
# We will apply these weights to each of the stocks in the relevant industry
industry_weights = (1 / len(industry_count)) / industry_count
industry_weights
```

```python
# Here is a double check to see our industry's each have the same weight. 
# So we will invest 14.2% of our money in each industry
industry_weights * industry_count
```

```python
# Using the merge function we can do SQL type joins in pandas
weights = constituents[['Share code', 'Share name', 'Industry']].merge(
    industry_weights,
    left_on='Industry',
    right_index=True
)

weights = weights[['Share code', 'Weight']]

weights['Share code'] = ('JSE:' + weights['Share code'])
weights.set_index('Share code', inplace=True)
weights = weights.loc[weights.index.isin(close.columns) == True]
weights.sort_index(inplace=True)

weights.head()
```

```python
# We will use the Satrix 40 as a proxy for the market.
market = web.DataReader('JSE:STX40', 'google', '2015-01-01', '2015-12-31')['Close']
```

```python
# Calculate the portfolio return by calculating the dot product of our prices and weights
portfolio = close.apply(lambda x: np.dot(x, weights.values[:, 0]), axis=1)
```

```python
# Let's rebase our portfolio and the market proxy and plot them to see who the winner is
data = {
    'portfolio': portfolio.div(portfolio[0]),
    'market': market.div(market[0])
}

data = DataFrame(data)

data.plot(
    title='Our portfolio vs the Top 40 (Proxy)',
    figsize=(12, 7)
)
```
